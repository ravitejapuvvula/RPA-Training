### Step 4. How many products cost more than $10.00?

```python
func= lambda x : float(x[1:-1])
chipo["item_price"]=chipo["item_price"].apply(func)
chipo[chipo["item_price"]>10.00]
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
price=chipo[["item_name","item_price","quantity"]]
price=price[price["quantity"]==1]
price=price.drop_duplicates(subset=['item_name'])

func= lambda x : float(x[1:-1])
price["item_price"]=price["item_price"].apply(func)
price.sort_values(by="item_price",ascending=False,inplace=True)
price[price.item_price>8.9]
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by="item_name",ascending=True,inplace=True)
chipo
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
expense=chipo.sort_values("item_price",ascending=False).head(1)
expense
```

Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie=chipo[chipo["item_name"]=="Veggie Salad Bowl"]
veggie.shape[0]
```
